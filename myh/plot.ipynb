```python
import matplotlib.pyplot as plt

# 创建一个figure对象和三个子图（3行1列），共享x轴
fig, axs = plt.subplots(3, 1, figsize=(10, 8), sharex=True)

# 第一张子图
axs[0].plot([1, 2, 3, 4], [10, 20, 25, 30], 'r')
axs[0].set_ylabel('Y轴 1')
axs[0].set_title('子图 1')

# 第二张子图
axs[1].plot([1, 2, 3, 4], [1, 4, 9, 16], 'g')
axs[1].set_ylabel('Y轴 2')
axs[1].set_title('子图 2')

# 第三张子图
axs[2].plot([1, 2, 3, 4], [0.5, 1.5, 2.5, 3.5], 'b')
axs[2].set_ylabel('Y轴 3')
axs[2].set_title('子图 3')
axs[2].set_xlabel('X轴')

# 调整布局
plt.tight_layout()

# 显示图形
plt.show()
```
